```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris, load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.metrics import accuracy_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Load example datasets
iris = load_iris()
X_clf, y_clf = iris.data, iris.target

print(iris.DESCR)

boston = load_boston()
X_reg, y_reg = boston.data, boston.target

print(boston.DESCR)


```

```python
# Split the data
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

```

```python
# Create and train decision tree classifiers
clf_tree = DecisionTreeClassifier(random_state=42)
clf_tree.fit(X_clf_train, y_clf_train)

```

```python
# Create and train decision tree regressors
reg_tree = DecisionTreeRegressor(random_state=42, min_samples_leaf=50)
reg_tree.fit(X_reg_train, y_reg_train)

```

```python
# Plot decision tree for classification
plt.figure(figsize=(10, 6))
plot_tree(clf_tree, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Classifier")
plt.show()

```

```python
# Plot decision tree for regression
plt.figure(figsize=(20, 16))
plot_tree(reg_tree, filled=True, feature_names=boston.feature_names)
plt.title("Decision Tree Regressor")
plt.show()
```
