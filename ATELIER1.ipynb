```python
import pandas as pd

# Charger vos propres données depuis un fichier CSV (par exemple)
# Assurez-vous de remplacer 'votre_chemin_vers_le_fichier.csv' par le chemin réel de votre fichier
df_external = pd.read_csv('creditcard.csv')

# Définir X (caractéristiques) et y (variable cible) à partir de vos propres données
# Assurez-vous d'adapter les noms des colonnes et les sélections selon votre structure de données
X = df_external.drop(columns=['Class'])  # Exclure la colonne cible pour former X
y = df_external['Class']  # Sélectionner la colonne cible comme y
```

```python
X.head()
```

```python
#Ajouter...
df = X.join(pd.Series(y, name='Class'))
df.head()
```

```python
#Codage...
le = LabelEncoder()
y = le.fit_transform(df['Class'])
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
X_lda = lda.fit_transform(X, y)
```

```python
lda.explained_variance_ratio_# les valeurs propres
```

```python
# Utilisez maintenant X_sampled et y_sampled pour votre analyse ou votre modélisation
plt.figure()
plt.xlabel('LD1')
plt.ylabel('LD1')  # Assurez-vous de spécifier la bonne dimension ici (LD1)
plt.scatter(X_lda[:, 0],y, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('LDA Projection')
plt.show()
```

```python


# Tracer le nuage de points avec la première composante discriminante
plt.figure()
plt.xlabel('LD1')
plt.ylabel('LD1')  # Assurez-vous de spécifier la bonne dimension ici (LD1)
plt.scatter(X_lda[:, 0], X_lda[:, 0], c=y, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('LDA Projection')
plt.show()
```

```python
# Select rows with Class = 0
class_0 =  df_external[ df_external['Class'] == 0].sample(n=492, random_state=42)

# Select rows with Class = 1
class_1 =  df_external[ df_external['Class'] == 1].sample(n=492, random_state=42)

# Concatenate both selected rows into a new DataFrame
new_data = pd.concat([class_0, class_1])
```

```python
X1 = new_data.drop(["Time","Amount","Class"],axis=1)
#Variable...
y1 = new_data["Class"]
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
X_lda = lda.fit_transform(X1, y1)
```

```python
lda.explained_variance_ratio_# les valeurs propres
```

```python
# Utilisez maintenant X_sampled et y_sampled pour votre analyse ou votre modélisation
plt.figure()
plt.xlabel('LD1')
plt.ylabel('LD1')  # Assurez-vous de spécifier la bonne dimension ici (LD1)
plt.scatter(X_lda[:, 0],y1, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('LDA Projection')
plt.show()
```
